```python
import numpy as np
from sklearn.linear_model import LinearRegression
horas = np.array([[8], [9], [10], [11], [12]])
temperatura  = np.array([12, 14, 25, 34, 29])

armazem=LinearRegression()
armazem.fit(horas, temperatura)

prev =[[13]]
precisao=armazem.predict(prev)
print(precisao)

```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

horas_estudo = np.array([[1], [3], [5], [6], [8]])
nota = np.array([6, 8, 9, 7, 9])

modelo = LinearRegression()
modelo.fit(horas_estudo, nota)

prev = [[0.5]]
precisao = modelo.predict(prev)
print("nota de previsão: ", precisao)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

tamanho = np.array([[7], [16]])
preço = np.array([70.000, 160.000])

modelo = LinearRegression()
modelo.fit(tamanho, preço)

prev = [[12]]
precisao = modelo.predict(prev)
print("preço previsto: ", precisao)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

anuncios = np.array([[10], [30], [60]])
vendas = np.array([100, 200, 500])

fds = LinearRegression()
fds.fit(anuncios, vendas)

prev = [[25]]
precisao = fds.predict(prev)
print("valores previstos: ", precisao)
```
