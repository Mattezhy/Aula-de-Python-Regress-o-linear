```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
texto = "O PLN nada mais é que o processamento de linguagem natural que processa qualquer idioma usado por seres naturais."
tokenização_palavra = word_tokenize(texto)
tokenização_sentença = sent_tokenize(texto)
```
